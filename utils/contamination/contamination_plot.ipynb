# PolitiFact

```python
import json
import matplotlib.pyplot as plt
import numpy as np

# --- File paths ---
llama_direct_politifact = 'contamination_direct_question_analysis_llama_politifact.json'
gemma_direct_politifact = 'contamination_direct_question_analysis_gemma_politifact.json'

# --- Helper to load contamination rate (%) ---
def get_rate(path, key='overall_analysis', field='contamination_rate'):
    with open(path) as f:
        d = json.load(f)
    return float(d[key][field])

# --- PolitiFact ---
llama_p = get_rate(llama_direct_politifact)
gemma_p = get_rate(gemma_direct_politifact)

# --- Plotting ---
def plot_direct(dataset, llama, gemma, filename=None):
    labels = ['Llama', 'Gemma']
    values = [llama, gemma]
    x = np.arange(len(labels))
    plt.figure(figsize=(5,5))
    bars = plt.bar(x, values, color=['#1f77b4','#ff7f0e'])
    plt.xticks(x, labels)
    plt.ylabel('Contamination Rate (%)')
    plt.title(dataset)
    for bar in bars:
        yval = bar.get_height()
        plt.text(bar.get_x() + bar.get_width()/2, yval + 0.1, f'{yval:.2f}%', ha='center', va='bottom', fontsize=12)
    plt.ylim(0, int(max(values)*1.2) + 1)
    plt.tight_layout()
    if filename:
        plt.savefig(filename, dpi=400)
    plt.show()

# --- Plot for PolitiFact ---
plot_direct('PolitiFact', llama_p, gemma_p, 'politifact_contamination_rate.png')
```
