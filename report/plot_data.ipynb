# Plot graph information

## KNN Graph

```python
import os

# read from the graph folder: ../graph/kdd2020/
graph_folder = '../graph/kdd2020/'

graph_infos = []    # list of graph_dict

for file in os.listdir(graph_folder):
    if file.endswith('.txt'):
        if 'knn' not in file:
            continue
        # read the file line by line
        with open(graph_folder + file, 'r') as f:
            lines = f.readlines()
            graph_dict = {}
            graph_dict['k'] = file.replace('.txt', '').split('_')[-1]
            for line in lines:
                try:
                    key, value = line.split(': ')
                    graph_dict[key] = int(value)
                except:
                    continue
            graph_infos.append(graph_dict)
print(graph_infos)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))

# Extract x and y values
x = [int(graph_dict['k']) for graph_dict in graph_infos]
y = [graph_dict['Total edges'] for graph_dict in graph_infos]

# Create the plot
plt.scatter(x, y)

# Annotate each point with its value
for i, txt in enumerate(y):
    plt.annotate(txt, (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xlabel('k')
plt.xticks(x)
plt.ylabel('Edges')
plt.title('KNN graph')

# Show the plot
plt.show()
```

# Plot Test Accuracy

## KNN Test Accuracy

```python
import os

# read from the graph folder: ../weights/
results_folder = '../weights/'

results_infos = []    # list of results_dict

for file in os.listdir(results_folder):
    if file.endswith('.txt'):
        if 'knn' not in file:
            continue
        if 'train_all' in file: # only read the few-shot results
            continue
        # read the file line by line
        with open(results_folder + file, 'r') as f:
            lines = f.readlines()
            # Test Accuracy (with best model): 0.6673346693386774
            result_dict = {}
            result_dict['k'] = file.replace('_results.txt', '').split('_')[-1]
            for line in lines:
                try:
                    key, value = line.split(': ')
                    result_dict[key] = float(value)
                except:
                    continue
            results_infos.append(result_dict)
print(results_infos)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(14, 5))

# Extract x and y values
x = [int(results_info['k']) for results_info in results_infos]
y = [results_info['Test Accuracy (with best model)'] for results_info in results_infos]
print(y)
# Create the plot
plt.scatter(x, y)

# Annotate each point with its value
for i, txt in enumerate(y):
    plt.annotate(round(txt, 2), (x[i], y[i]), textcoords="offset points", xytext=(0,10), ha='center')

# Add labels and title
plt.xticks(x)
plt.ylim(min(y)-0.01, max(y)+0.01)
plt.xlabel('k')
plt.ylabel('Test Accuracy')
plt.title('KNN graph')

# Show the plot
plt.show()
```
